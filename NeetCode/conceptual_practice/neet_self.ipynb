Contains Duplicate
Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:
Input: nums = [1, 2, 3, 3]
Output: true

Example 2:
Input: nums = [1, 2, 3, 4]
Output: false

```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        x = set()
        for m in nums:
            if m in x:
                return True
            else:
                x.add(m)
        return False
sol = Solution()
y = sol.hasDuplicate(nums = [1,2,3,3])
print(y)
```

Two Sum
Solved 
Given an array of integers nums and an integer target, return the indices i and j such that nums[i] + nums[j] == target and i != j.

You may assume that every input has exactly one pair of indices i and j that satisfy the condition.

Return the answer with the smaller index first.

Example 1:

Input: 
nums = [3,4,5,6], target = 7

Output: [0,1]
Explanation: nums[0] + nums[1] == 7, so we return [0, 1].

Example 2:

Input: nums = [4,5,6], target = 10

Output: [0,2]
Example 3:

Input: nums = [5,5], target = 10

Output: [0,1]

```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(len(nums)-1,i,-1):
                if nums[i] == target - nums[j]:
                    return(i,j)
        return(0,0)
    
sol = Solution()
x=sol.twoSum(nums = [3,4,5,6], target = 7)
print(x)
```

```python
# now O(n)
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i, num in enumerate(nums):
            need = target - num
            if need in seen:
                return [seen[need], i]
            seen[num] = i

sol = Solution()
x=sol.twoSum(nums = [3,4,5,6], target = 7)
print(x)
```

```python
# conceptual understanding of defaultdict:
from typing import List
from collections import defaultdict
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        groups = defaultdict(int)
        for n in nums:
            if n % 2 == 0:
                groups["even"] += 1
            else:
                groups["odd"] += 1
        return dict(groups)

sol = Solution()
x=sol.topKFrequent(nums = [1,2,2,3,3,3], k = 2)
print(x)
```

```python
# above one was O(nlogn) in a way here's one with bucket sort thats linear time:
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = {}
        freq = [[] for i in range(len(nums) + 1)]

        for num in nums:
            count[num] = 1 + count.get(num, 0)
        for num, cnt in count.items():
            freq[cnt].append(num)

        res = []
        for i in range(len(freq) - 1, 0, -1):
            for num in freq[i]:
                res.append(num)
                if len(res) == k:
                    return res
```
