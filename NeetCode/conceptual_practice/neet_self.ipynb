Contains Duplicate
Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:
Input: nums = [1, 2, 3, 3]
Output: true

Example 2:
Input: nums = [1, 2, 3, 4]
Output: false

```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        x = set()
        for m in nums:
            if m in x:
                return True
            else:
                x.add(m)
        return False
sol = Solution()
y = sol.hasDuplicate(nums = [1,2,3,3])
print(y)
```
